```python
# 各科目の点数を変数として定義
japanese = 85  # 国語
math = 90      # 数学
english = 75   # 英語

# 平均点を計算
average = (japanese + math + english) / 3

# 結果を表示
print("3科目の平均点は", average, "点です")
```
