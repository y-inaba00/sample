```python
#繰り返し　for文
for i in range(11):
    print(i)
```

```python
#繰り返し　range関数
for i in range(2, 21, 2):
    print(i)
```

```python
#繰り返し　while文
i = 0
while i < 10:
    i += 1
    print(i)

print("処理を終了します")
```

```python
#繰り返し処理の制御
#break: ループを途中で終了します。

i = 0
while True:
    i += 1
    if i > 10:
        break
    print(i)

print("処理を終了します")
```

```python
#繰り返し処理の制御
#continue: 現在の処理をスキップして次のループに進みます。

i = 0
while True:
    i += 1
    if i > 10:
        break
    if i % 2 == 1:
        continue
    print(i)
```

```python
score = int(input("点数を入力してください: "))

if score >= 60:
    print("合格です")

print("処理を終了します")
```

```python
#例外処理
try:
    #エラーが起こる可能性のある処理
    num = int(input("整数を入力してください:"))
    result = 100 / num
    print(f"100 を {num} で割った結果は {result} です")

except ZeroDivisionError:
    #0で割った場合のエラー処理
    print("エラー:0で割ることはできません")

except ValueError:
    #数値以外を入力した場合のエラー処理
    print("エラー:整数を入力してください")
```

```python
#例外処理2
try:
    num = int(input("整数を入力してください:"))
    result = 100 / num

except ZeroDivisionError:
    print("エラー: 0 で割ることはできません")
except ValueError:
    print("エラー: 整数を入力してください")

else:
    print(f"計算結果: {result}")

finally:
    print("プログラム終了")
```

```python
#シーケンス
s = "abcdefghijklmnopqrstuvwxyz"

print(s[0])
print(s[-1])

print(s[0:7])
print(s[-5:])
print(s[::2]) #2個おきに表示
```

```python
#反復処理
s = "TechAcademy"

for char in s:
    print(char.upper()) #各文字を大文字に変換して表示
```

```python
#長さの取得
s = "TechAcademy"
print(len(s))

if len(s) > 10:
    print("長い文字列です。")
```

```python
#回文判定
s = "にんてんどううどんてんに"

#形式: s[開始位置:終了位置:増分]
#開始・終了位置は省略可能
#増分を負にすると逆順に取得できます。
if s == s[::-1]:
    print(f"「{s}」は回文です")

else:
    print(f"「{s}」は回文ではありません")
```

```python
#リスト
numbers = [1, 2, 3, 4, 5]
fruits = ["apple", "banana", "cherry"]
mixed = [1, "apple", 3.14, True]
empty_list = []

print(numbers, fruits, mixed, empty_list)
```

```python
#list()関数

string_to_list = list("hello")
tuple_to_list = list((1, 2, 3))
range_to_list = list(range(5))

print(string_to_list, tuple_to_list, range_to_list)
```

```python
none_list = [None] * 5
zeros = [0] * 10

print(none_list, zeros)
```

```python
original = [1, 2, 3]
copy1 = original[:]
copy2 = list(original)
copy3 = original.copy()

print(copy1, copy2, copy3)
```
