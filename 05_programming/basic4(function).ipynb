```python
#組み込み関数

#print()
print("Hello, Python!")

print("Test", end="") #改行しない
print("Next line")  

print(end= "\n")
#len()
print("文字列のの長さ:", len("Python")) 
print("リストの要素数:", len([1, 2, 3, 4]))

print(end= "\n")
#type()
print(type(123))
print(type("Python"))   
print(type([1, 2, 3]))

print(end= "\n")
#max() min()
print("MAX:", max(5, 10, 20))
print("MIN:", min([1, 2, 3, 4]))

print(end= "\n")
#sum()
#sumはまとめたデータを受け取るので、リストやタプルにする
print("合計:", sum([1, 2, 3 ,4]))
print("合計:", sum((10, 20, 30)))

print(end= "\n")
#range()
#指定範囲の数値を生成
print("range生成")
for i in range(5):
    print(i)

print(end= "\n")
print(list(range(1, 10, 2))) #1から9を2個飛ばしで
```

```python
#関数の定義

# def 関数名(引数1, 引数2, ...):
    # 関数の処理
# return 戻り値

#回文判定を関数にまとめる
def is_palindrome(s): #is-は「-かどうかを判定する」
    if s == s[::-1]:
        print(f"「{s}」は回文です")
    else:
        print(f"「{s}」は回文ではありません")

is_palindrome("にんてんどううどんてんに")
```

```python
#引数
def greet(name):
    print(f"こんにちは、{name}さん。")

greet("太郎")


print(end= "\n")
#複数の引数
def introduce(name, age):
    print(f"こんにちは、{name}さん！年齢は{age}歳ですね。")

introduce("花子", 25)

print(end= "\n")
#デフォルト引数
def introduce(name, age=18):
    print(f"こんにちは、{name}さん！年齢は{age}歳ですね。")

introduce("太郎")

print(end= "\n")
#可変長引数
#引数の数が決まっていない場合に使う

#*args: 複数の位置引数を受け取る
def add(*args):
    print("合計:", sum(args))

add(1, 2, 3, 4)


print(end= "\n")
#**kwargs: 複数のキーワード引数を受け取る
def total(**kwargs):
    print(kwargs)

total(apple=1, banana=2)


print(end= "\n")
#戻り値
def add(a, b):
    return a + b

result = add(3, 5)
print(result)


#複数の戻り値
def calculate(a, b):
    return a + b, a * b

sum_result, product_result = calculate(3, 4)
print("足し算:", sum_result)
print("掛け算:", product_result)
```

```python
#モジュールとライブラリの活用

#基本のインポート
import math
print(math.sqrt(16)) #「sqrt(16)」16の平方根（ルート）を求める


print(end= "\n")
#エイリアスを使う
import datetime as dt #モジュールにdtという別名をつけて読み込む
now = dt.datetime.now()　#「datetime.datetime.now()」でモジュール名とクラス名が重なっている
print("現在の日時:", now)
```

```python
#基準ライブラリの活用

#math モジュール
print(math.sqrt(25))  
print(math.exp(1))    
print(math.pi) 

print(end= "\n")
#datetime モジュール
now = dt.datetime.now()
print(now)

print(end= "\n")
birthday = dt.datetime(1995, 5, 17)
print("誕生日:", birthday)

print(end= "\n")
date1 = dt.datetime(2023, 1, 1)
date2 = dt.datetime(2024, 1, 1)
print("日数差:", (date2 - date1).days, "日")  # 日数差

print(end= "\n")
#random モジュール
import random

print(random.randint(1, 10))  # 1〜10のランダムな整数

my_list = [1, 2, 3, 4, 5]
random.shuffle(my_list)       # リストをランダムに並び替え

print(my_list)
print(random.choice(my_list)) # ランダムな要素を1つ選択
```
